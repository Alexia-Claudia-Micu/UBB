```python
import random

def spinner(probabilityDistribution):
    rand_num = random.random()
    cumulative_prob = 0
    for i in range(len(probabilityDistribution)):
        cumulative_prob += probabilityDistribution[i]
        if rand_num < cumulative_prob:
            return i
    return len(probabilityDistribution) - 1

probabilityDistribution = [0.2, 0.3, 0.1, 0.4]
index = spinner(probabilityDistribution)
print(index)
```

```python
import random

def spinnerAngles(probabilityDistribution):
    total_degrees = sum(probabilityDistribution)
    rand_degrees = random.uniform(0, total_degrees)
    cumulative_degrees = 0
    for i, angle in enumerate(probabilityDistribution):
        cumulative_degrees += angle
        if rand_degrees < cumulative_degrees:
            return i
    return len(probabilityDistribution) - 1

probabilityDistribution = [30,45,15,60]
index = spinnerAngles(probabilityDistribution)
print(index)
```

```python
import random

def monte_carlo_simulation(num_points):
    R = 30  # Radius of the pipe (in cm)
    l = 100  # Length of the square yard (in cm)
    r = 1  # Radius of the balls (in cm)

    num_points_inside_circle = 0

    for _ in range(num_points):
        x = random.uniform(-l/2, l/2)
        y = random.uniform(-l/2, l/2)

        if x**2 + y**2 <= R**2:
            num_points_inside_circle += 1

    probability = num_points_inside_circle / num_points
    return probability

# Perform the Monte Carlo simulation with 10^6 points
num_points = 10**6
estimated_probability = monte_carlo_simulation(num_points)
print("Estimated probability using Monte Carlo simulation:", estimated_probability)
```
